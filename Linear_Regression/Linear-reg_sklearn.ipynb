```python
## Making the necessary imports
import pandas
import numpy as np
import matplotlib.pyplot as pt
import seaborn
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
## Loading the data set
salary_data = pandas.read_csv("./Salary_Data.csv")
```

```python
## Analyzing the data 
print("Your data has the following mathematical features:")
print(salary_data.describe())
```

```python
## Plotting the data 
pt.title("Salary data distribution")
seaborn.histplot(salary_data['Salary'], kde=True)
```

```python
## Plotting the graph

pt.scatter(salary_data['YearsExperience'], salary_data['Salary'])
pt.title("Exp vs Salary")
pt.xlabel("Years")
pt.ylabel("Salary")
pt.show()
```

```python
## Defining and spliting the data

x = salary_data.iloc[:, :1] # Years {Independent vars}
y = salary_data.iloc[:, 1:] # Salaries {Dependent vars}

# Splitting the data

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
## Training the rady made model, actually it should be fitting the data on a model

regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
## Prediction test

y_pred_train = regressor.predict(x_train)
y_pred_test = regressor.predict(x_test)

```

```python
## Plotting the Train Set

pt.scatter(x_train, y_train)
pt.plot(x_train, y_pred_train)
pt.title('Salary vs Experience (Training Set)')
pt.xlabel('Years of Experience')
pt.ylabel('Salary')
pt.legend(['X_train/y_train', 'X_train/Pred(y_test)'], title = 'Sal/Exp', loc='best', facecolor='white')
pt.box(False)
pt.show()
```

```python
## Plotting the Train Set

pt.scatter(x_train, y_train, color = 'red')
pt.plot(x_train, y_pred_train, color="red")
pt.title('Salary vs Experience (Training Set)')
pt.xlabel('Years of Experience')
pt.ylabel('Salary')
pt.legend(['X_train/y_train', 'X_train/Pred(y_test)'], title = 'Sal/Exp', loc='best', facecolor='white')
pt.box(False)
pt.show()
```
